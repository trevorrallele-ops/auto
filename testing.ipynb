```python
import requests
import pandas as pd

# 1Ô∏è‚É£ Your Alpha Vantage API key
API_KEY = "YOUR_ALPHA_VANTAGE_KEY"  # Replace with your real key
symbol = "GLD"

# 2Ô∏è‚É£ API endpoint for weekly data
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol={symbol}&apikey={API_KEY}'

# 3Ô∏è‚É£ Make the request
response = requests.get(url)
data = response.json()

# 4Ô∏è‚É£ Check for errors
if "Weekly Adjusted Time Series" not in data:
    print("Error fetching data:", data)
    exit()

# 5Ô∏è‚É£ Convert to DataFrame
ts_weekly = data["Weekly Adjusted Time Series"]
df = pd.DataFrame.from_dict(ts_weekly, orient='index')

# 6Ô∏è‚É£ Rename columns
df.rename(columns={
    "1. open": "Open",
    "2. high": "High",
    "3. low": "Low",
    "4. close": "Close",
    "5. adjusted close": "Adj Close",
    "6. volume": "Volume",
    "7. dividend amount": "Dividend"
}, inplace=True)

# 7Ô∏è‚É£ Convert index to datetime
df.index = pd.to_datetime(df.index)

# 8Ô∏è‚É£ Convert numeric columns to float
for col in ["Open", "High", "Low", "Close", "Adj Close", "Volume", "Dividend"]:
    df[col] = df[col].astype(float)

# 9Ô∏è‚É£ Sort by date
df.sort_index(inplace=True)

# üîü Print head and tail
print(df.head())
print(df.tail())

# 1Ô∏è‚É£1Ô∏è‚É£ Optional: save to CSV
df.to_csv("GLD_weekly.csv")
```

```python
import yfinance as yf
import pandas as pd

# Ticker for GLD ETF
symbol = "GLD"
gld = yf.Ticker(symbol)

# Fetch full historical daily data
df = gld.history(period="max")  # or period="1y" for last year

# Optional: reset index to have Date as a column
df.reset_index(inplace=True)

# Save to CSV
df.to_csv("GLD_yf_daily.csv", index=False)

# Quick peek
print(df.head())
print(df.tail())
```

```python
import yfinance as yf
import pandas as pd

# Ticker for GLD ETF
symbol = "GLD"
gld = yf.Ticker(symbol)

# Fetch full historical weekly data
df = gld.history(period="max", interval="1wk")  # interval changed to 1wk

# Optional: reset index to have Date as a column
df.reset_index(inplace=True)

# Save to CSV
df.to_csv("GLD_yf_weekly.csv", index=False)

# Quick peek
print(df.head())
print(df.tail())
```
